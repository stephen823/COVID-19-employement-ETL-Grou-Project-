# 1. Pandemic Data

```python
# Dependencies and Setup
import pandas as pd
```

```python
from requests import get  
url="https://www.kaggle.com/ryanxjhan/coronaviruscovid19-canada/download"
file_name="./Data/coronaviruscovid19-canada.zip"
```

```python
#Extract data from the downloaded zip file
# download(url,file_name)
import zipfile
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall("./Data/Canada coronavirus data")
```

```python
#Define data path
cases_load="Data/Canada coronavirus data/cases.csv"
mortality_load="Data/Canada coronavirus data/mortality.csv"
recovered_load="Data/Canada coronavirus data/recovered.csv"
testing_load="Data/Canada coronavirus data/testing.csv"
```

```python
#Load data
cases = pd.read_csv(cases_load)
mortality = pd.read_csv(mortality_load)
recovered = pd.read_csv(recovered_load)
testing = pd.read_csv(testing_load)
```

```python
#Remove column 
cases1=cases[["case_id","provincial_case_id","age","sex","health_region","province","country","report_week","travel_yn","travel_history_country","locally_acquired"]]
mortality1=mortality[["death_id","province_death_id","case_id","age","sex","health_region","province", "country","date_death_report"]]
recovered1=recovered[["date_recovered", "province","cumulative_recovered","province_source"]]
testing1=testing[["date_testing","province","cumulative_testing","province_source"]]

recovered1
```

```python
#Remove missing data and replace them in with appropriate values
cases1.locally_acquired=cases1[["locally_acquired"]].fillna("Not Reported")
cases1.travel_history_country=cases1[["travel_history_country"]].fillna("Not Reported")

mortality1.case_id=mortality1[["case_id"]].fillna("Not Reported")

recovered1.cumulative_recovered=recovered1[["cumulative_recovered"]].fillna(0)
recovered1.province_source=recovered1[["province_source"]].fillna("Not Reported")

testing1.cumulative_testing=testing1[["cumulative_testing"]].fillna(0)
testing1.province_source=testing1[["province_source"]].fillna("Not Reported")



#
cases1["locally_acquired"] = cases1["locally_acquired"].replace({"Close contact":"Close Contact","close contact":"Close Contact"})
cases1["age"] = cases1["age"].replace({"<1":"0-10","2":"0-10","<10":"0-10"})
cases1["age"] = cases1["age"].replace({"10-19":"11-30","20-29":"11-30"})
cases1["age"] = cases1["age"].replace({"30-39":"31-50","40-49":"31-50","50":"31-50"})
cases1["age"] = cases1["age"].replace({"50-59":"above 50","60-69":"above 50","61":"above 50","70-79":"above 50","80-89":"above 50","90-99":"above 50"})

cases1["sex"] = cases1["sex"].replace({"Female":"Females","Male":"Males"})
mortality1["sex"] = mortality1["sex"].replace({"Female":"Females","Male":"Males"})

cases1.age=cases1.age.astype(str)
cases1.sex=cases1.sex.astype(str)
cases1.health_region=cases1.health_region.astype(str)
cases1.country=cases1.country.astype(str)
cases1.travel_history_country=cases1.travel_history_country.astype(str)
cases1.locally_acquired=cases1.locally_acquired.astype(str)
print(cases1.dtypes)
cases1
```

```python
from datetime import datetime
#Add Month column to table
cases1.report_week=pd.to_datetime(cases1.report_week)
cases1['date']=["date"]*len(cases1.index) 
cases1
for index, row in cases1.iterrows():
    month=row['report_week'].month
    year=row['report_week'].year
    cases1.at[index, "date"]=datetime(year,month,1)
    cases1.at[index, "date"]=cases1.at[index, "date"].strftime('%Y-%m-%d')
    
mortality1.date_death_report=pd.to_datetime(mortality1.date_death_report)
mortality1['date']=["date"]*len(mortality1.index) 
for index, row in mortality1.iterrows():
    month=row['date_death_report'].month
    year=row['date_death_report'].year
    mortality1.at[index, "date"]=datetime(year,month,1)
    mortality1.at[index, "date"]=mortality1.at[index, "date"].strftime('%Y-%m-%d')
    
    
recovered1.date_recovered=pd.to_datetime(recovered1.date_recovered)
recovered1['date']=["date"]*len(recovered1.index) 
cases1
for index, row in recovered1.iterrows():
    month=row['date_recovered'].month
    year=row['date_recovered'].year
    recovered1.at[index, "date"]=datetime(year,month,1)
    recovered1.at[index, "date"]=recovered1.at[index, "date"].strftime('%Y-%m-%d')
    counting=recovered1.at[index,"cumulative_recovered"]
    recovered1.at[index,"cumulative_recovered"]=int(counting)

recovered1.cumulative_recovered=recovered1.cumulative_recovered.astype(int)

testing1.date_testing=pd.to_datetime(testing1.date_testing)
testing1['date']=["date"]*len(testing1.index) 
cases1
for index, row in testing1.iterrows():
    month=row['date_testing'].month
    year=row['date_testing'].year
    testing1.at[index, "date"]=datetime(year,month,1)
    testing1.at[index, "date"]=testing1.at[index, "date"].strftime('%Y-%m-%d')
    

```

```python
print(cases1)
print(mortality1)
print(recovered1)
print(testing1)
import numpy as np
np.unique([v for v in cases1.province])
recovered1
```

```python
Alberta=cases1[cases1["province"]=="Alberta"]
BC=cases1[cases1["province"]=="BC"]
Manitoba=cases1[cases1["province"]=="Manitoba"]
NL=cases1[cases1["province"]=="NL"]
NWT=cases1[cases1["province"]=="NWT"]
New_Brunswick=cases1[cases1["province"]=="New Brunswick"]
Nova_Scotia=cases1[cases1["province"]=="Nova Scotia"]
PEI=cases1[cases1["province"]=="PEI"]
Quebec=cases1[cases1["province"]=="Quebec"]
Yukon=cases1[cases1["province"]=="Yukon"]
Repatriated=cases1[cases1["province"]=="Repatriated"]
testing1
```

```python
#Summarize cases table
cases11=cases1[["case_id","sex", "date"]]
cases_count=cases11.groupby(['sex',"date"]).count()
cases_count["date"]=[v[1] for v in cases_count.index]
cases_count["sex"]=[v[0] for v in cases_count.index]
cases_count=cases_count[["sex","date","case_id"]]
cases_count["category"]=["cases"]*len(cases_count.index)

cases_count.rename(columns={'case_id':'cases_count'}, inplace=True)

cases_count
```

```python
#Summarize mortality table
mortality11=mortality1[["case_id","sex", "date"]]
mortality_count=mortality11.groupby(["date","sex"]).count()
mortality_count["date"]=[v[0] for v in mortality_count.index]
mortality_count["sex"]=[v[1] for v in mortality_count.index]
mortality_count=mortality_count[["sex","date","case_id"]]
mortality_count["category"]=["mortality"]*len(mortality_count.index)

mortality_count.rename(columns={'case_id':'mortality_count'}, inplace=True)

mortality_count

```

```python
#Summarize recovered table
recovered11=recovered1[["date","province","cumulative_recovered"]]
recovered_count=recovered11.groupby(["province","date"]).last()
recovered_count["date"]=[v[1] for v in recovered_count.index]
recovered_count["province"]=[v[0] for v in recovered_count.index]
recovered_count=recovered_count[["province","date","cumulative_recovered"]]

recovered_count["category"]=["recovered"]*len(recovered_count.index)

recovered_count
```

```python
#Summarize testing table 
testing11=testing1[["date","province","cumulative_testing"]]
testing_count=testing11.groupby(["province","date"]).last()
testing_count["date"]=[v[1] for v in testing_count.index]
testing_count["province"]=[v[0] for v in testing_count.index]
testing_count=testing_count[["province","date","cumulative_testing"]]
testing_count["category"]=["testing"]*len(testing_count.index)
testing_count
```

```python
cases_count.to_csv("./Data/Cleaned Data/cases_count.csv",
                   index=False,
                   header=True
                  )

mortality_count.to_csv("./Data/Cleaned Data/mortality_count.csv",
                   index=False,
                   header=True
                  )

recovered_count.to_csv("./Data/Cleaned Data/recovered_count.csv",
                   index=False,
                   header=True
                  )

testing_count.to_csv("./Data/Cleaned Data/testing_count.csv",
                   index=False,
                   header=True
                  )
```

```python
import pandas as pd
```

```python
from requests import get  # to make GET request

url = 'https://www150.statcan.gc.ca/n1/en/tbl/csv/14100017-eng.zip?st=9vcqm52I'
file_name = './001_COVID_19_UNEMPLOYMENT/Unemployment_Rate_ZIP.zip'


def download(url, file_name):
    # open in binary mode
    with open(file_name, "wb") as file:
        # get request
        response = get(url)
        # write to file
        file.write(response.content)


download(url, file_name)
```

```python
import zipfile
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('./001_COVID_19_UNEMPLOYMENT')
```
