```python
import pandas as pd
```

```python
from requests import get  # to make GET request

url = 'https://www150.statcan.gc.ca/n1/en/tbl/csv/14100017-eng.zip?st=9vcqm52I'
file_name = './001_COVID_19_UNEMPLOYMENT/Unemployment_Rate_ZIP.zip'


def download(url, file_name):
    # open in binary mode
    with open(file_name, "wb") as file:
        # get request
        response = get(url)
        # write to file
        file.write(response.content)


download(url, file_name)
```

```python
import zipfile
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('./001_COVID_19_UNEMPLOYMENT')
```

```python
path = './001_COVID_19_UNEMPLOYMENT/14100017.csv'
DOWNLOADED_FILE_df = pd.read_csv(path)
```

```python
UNEMPLOYMENT_df = DOWNLOADED_FILE_df[['Sex', 'Age group', 'VALUE', 'REF_DATE', 'UOM']]
UNEMPLOYMENT_df.head()
```

```python
UNEMPLOYMENT_df["REF_DATE"] = UNEMPLOYMENT_df["REF_DATE"]+"-01"
UNEMPLOYMENT_df.head()
```

```python
UNEMPLOYMENT_df2 = UNEMPLOYMENT_df.loc[(UNEMPLOYMENT_df["REF_DATE"] == '2019-11-01')| 
                    (UNEMPLOYMENT_df["REF_DATE"] == '2019-12-01')| 
                    (UNEMPLOYMENT_df["REF_DATE"] == '2020-01-01')|
                    (UNEMPLOYMENT_df["REF_DATE"] == '2020-02-01')|
                    (UNEMPLOYMENT_df["REF_DATE"] == '2020-03-01'), :]
```
